```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```python
file_path ="/Volumes/workspace/default/kaggle_datasets/Global_Mobile_Prices_2025_Extended.csv"

Mobile_df = spark.read.format("csv").option("header",True).option("inferschema",True).load(file_path)
display(Mobile_df)
```

```python
Mobile_df.createOrReplaceTempView("Mobiles")
# it will remaned the kaggle data set
```

```python
%sql
select * from Mobiles;
```

```python
Mobile_df.columns
```

```python
%sql
SELECT
brand,
count(price_usd) as total_count,
sum(price_usd) as sum,
max(price_usd) as maximum,
min(price_usd) as minimum,
avg(price_usd) as average
from Mobiles
where brand like '%le' or brand like '%mi%'
GROUP by brand;
```

```python
%sql
-- # window Functions
-- # row number
-- # rank
-- # dense rank
-- # percent rank
select distinct brand,
row_number() over (PARTITION BY brand ORDER BY price_usd desc) as rownumber
from Mobiles
limit 10;

```

```python
%sql
SELECT
    DENSE_RANK() OVER (ORDER BY price_usd ASC) AS rank_price,
    brand,
    model,
    price_usd
FROM Mobiles;
```
