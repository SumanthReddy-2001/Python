```python
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType, StringType, StructField, StructType
```

```python
# without strict schema

row_data = [
    (10,'Sumanth',20000,'Bangalore'),
    (11,'Diwakar',10000,'Chennai'),
    (12,'Kumar',33000,'Hyderabad'),
    (13,'Reddy',40000,'Mysore')
]

columns = ['id','name','salary','city']

emp_df = spark.createDataFrame(data=row_data,schema=columns)
display(emp_df)
```

```python
emp_df.show()
```

```python
# with strict schema

row_data1 = [
    (10,'Sumanth',20000,'Bangalore'),
    (11,'Diwakar',10000,'Chennai'),
    (12,'Kumar',33000,'Hyderabad'),
    (13,'Reddy',40000,'Mysore')
]

columns = StructField
[
        StructField('id',IntegerType(),True),
        StructField('name',StringType(),True),
        StructField('salary',IntegerType(),True),
        StructField('city',StringType(),True)
]

emp_df1 = spark.createDataFrame(data=row_data1,schema=columns)
display(emp_df1)
```

```python
file_path = "/Volumes/workspace/default/kaggle_datasets/Latest 2025 movies Datasets.csv"

Movies_df = spark.read.format("csv").load(file_path)
display(Movies_df)
```

```python
Movies_df = spark.read.format("csv").option("header","true").load(file_path)
display(Movies_df)
```

```python
Movies_df = spark.read.format("csv").option("header","true").option("inferSchema","true").load(file_path)
display(Movies_df)
```

```python
#Filter the data frame to get only movies released in
# we need to over ride the data frame with in its own dataframe  ex : movies = movies.filter()
Movies_df= Movies_df.filter(F.col("original_language")=="hi")
display(Movies_df)
```

```python
# using Select Function
# it will select the columns which are specified in the select function
Movies_df = Movies_df.select(
    F.col("title"),
    F.col("original_language")
)
display(Movies_df)
```
